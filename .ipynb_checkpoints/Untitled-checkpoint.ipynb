```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import tensorflow as tf
import time
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder,LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split,StratifiedKFold
from sklearn import metrics
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from sklearn.metrics import confusion_matrix, roc_auc_score, mean_absolute_error
from pandas_profiling import ProfileReport
from xgboost import XGBClassifier
from tensorflow.python.client import device_lib
```

```python
df = pd.read_csv('./dataset/train.csv')
df
```

```python
df.loc[df.Gender == 'Male','Gender'] = 1
df.loc[df.Gender == 'Female','Gender'] = 0
df.loc[df.Vehicle_Age == '> 2 Years','Vehicle_Age'] = 2
df.loc[df.Vehicle_Age == '< 1 Year','Vehicle_Age'] = 1.5
df.loc[df.Vehicle_Age == '1-2 Year','Vehicle_Age'] = 0.5
df.loc[df.Vehicle_Damage == 'Yes','Vehicle_Damage'] = 1
df.loc[df.Vehicle_Damage == 'No','Vehicle_Damage'] = 0
Scaler = StandardScaler()
df[['Age','Region_Code','Vehicle_Age','Annual_Premium','Policy_Sales_Channel','Vintage']]= Scaler.fit_transform(df[['Age','Region_Code','Vehicle_Age','Annual_Premium','Policy_Sales_Channel','Vintage']])
df_cleaned = df
df_cleaned.to_csv('./dataset/train_cleaned.csv',index=False)
```

```python
df_train = pd.read_csv('./dataset/train_cleaned.csv')
```

```python
df_positive = df_train[df_train['Response']==1]
df_negative = df_train[0:170000][df_train['Response']==0]
df_train = pd.concat([df_negative,df_positive])
df_train
```

```python
df_train.drop('id',axis=1, inplace=True)
X=df_train.drop('Response',axis=1)
y = df_train['Response']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, train_size = 0.8, random_state=10)
```

```python
with tf.device('/GPU:0'):
    model = Sequential()
    model.add(Dense(32, input_dim=X.shape[1], kernel_initializer='uniform',activation="relu" ))
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(0.3))
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(0.3))
    model.add(Dense(units = 1, activation='sigmoid', kernel_initializer='uniform'))
    model.compile( metrics = [tf.keras.metrics.BinaryAccuracy()],loss = 'binary_crossentropy' )
```

```python
skf = StratifiedKFold(n_splits=5)
for i,(train_index, test_index) in enumerate(skf.split(X, y)):
    
    start_time = time.time()
    print('Running Fold: '+ str(i+1))
    model.fit(X.iloc[train_index], y.iloc[train_index], verbose=0, epochs=50, batch_size=264, )
    y_pred = model.predict(X.iloc[test_index])
    rounded = y_pred>0.5
    print("Accuracy:",metrics.accuracy_score(y.iloc[test_index], rounded))
    print("Precision:",metrics.precision_score(y.iloc[test_index], rounded))
    print("Recall:",metrics.recall_score(y.iloc[test_index], rounded))
    print("AUC_Score:",roc_auc_score(y.iloc[test_index], y_pred))
    model.save('./ML_Model/loan_nn_'+str(round(metrics.accuracy_score(y.iloc[test_index], rounded),ndigits=2))+'_'+str(round(roc_auc_score(y.iloc[test_index], y_pred),ndigits=4))+'_'+time.strftime('%H-%M-%S'))
    print("Training Time: " + str(round((time.time() - start_time),ndigits=2)))
```

```python
model.fit(X_train, y_train, verbose=1, epochs=40, batch_size=264 )
```

```python
y_pred = model.predict(X_test)
#rounded = y_pred>0.6
rounded = np.round(y_pred)

print("Accuracy:",metrics.accuracy_score(y_test, rounded))
print("Precision:",metrics.precision_score(y_test, rounded))
print("Recall:",metrics.recall_score(y_test, rounded))
print("AUC_Score:",roc_auc_score(y_test, y_pred))
```
